```python
print(__doc__)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
def underline_key(phrase_u, phrase_nu):
    sp0, sp1 = "\033[1;;4m", "\033[1;;2m" 
    text_u = sp0+phrase_u+sp1+" "+phrase_nu
    print(text_u)
```

```python
df_data = pd.read_csv("data/data-breast-cancer-wisconsin-QueryResult.csv")
df_data.head()
```

```python
print(list(df_data.columns.values))
df_data.dtypes
```

```python
#MODIFY THE FOLLIWING TWO LINES FOR DIFFERENT DATA SET
df_data_x = df_data.iloc[:,3:]
df_data_y = df_data.iloc[:,2]
# test/training x/y variables used for multiple parts of code
X_train, X_test, y_train, y_test = train_test_split(df_data_x, df_data_y, random_state = 0)
underline_key("X_train shape:",str(X_train.shape))
underline_key("y_train shape:",str(y_train.shape))
underline_key("X_test shape:",str(X_test.shape))
underline_key("y_test shape:",str(y_test.shape))
```

```python
feature_names = list(df_data.columns.values[3:])
class_names = list(df_data.diagnosis.unique())
```

```python

import pickle
with open('saved_variables/X_y_train_test', 'wb') as open_pickle_file:
    pickle.dump([X_train, X_test, y_train, y_test], open_pickle_file)
    
with open('saved_variables/df_X_y_data', 'wb') as open_pickle_file:
    pickle.dump([df_data_x, df_data_y, df_data], open_pickle_file)
    
with open('saved_variables/feature_class_names', 'wb') as open_pickle_file:
    pickle.dump([feature_names, class_names], open_pickle_file)
```
